# Group Similar Colors Together

```python
import numpy as np

#(0, 0, 0) - black
#(127, 127, 127) - gray
#(136, 0, 21) - burgundy
#(237, 28, 36) - red
#(255, 127, 39) - orange
#(255, 242, 0) - yellow
#(34, 177, 76) - green
#(203, 228, 253) - blue
#(0, 162, 232) - dark blue
#(63, 72, 204) - purple
#(255, 255, 255) - white
#(195, 195, 195) - light gray
#(185, 122, 87) - light brown
#(255, 174, 201) - light pink
#(255, 201, 14) - dark yellow
#(239, 228, 176) - light yellow
#(181, 230, 29) - light green
#(153, 217, 234) - light blue
#(112, 146, 190) - dark blue
#(200, 191, 231) - light purple
#(40, 40, 40) - dark gray
#(149, 69, 53) - chestnut
#(92, 64, 51) - dark brown
#(123, 63, 0) - chocolate
#(54, 69, 79) - charcoal
#(105, 105, 105) - dim gray
#(112, 128, 144) - slate gray
#(114, 116, 114) - nickel

list_of_colors = [[0,0,0],[127,127,127],[136,0,21],[237,28,36],[255,127,39], [255,242,0],[34,177,76],[203,228,253],[0,162,232], 
                  [63,72,204], [255,255,255],[195,195,195],[185,122,87],[255,174,201],[255,201,14],[255,201,14],[239,228,176],
                  [181,230,29], [153,217,234], [112,146,190],[200,191,231], [40,40,40], [149,69,53],[92,64,51], [123,63,0], 
                  [54,69,79], [105,105,105], [112,128,148], [114,116,114]]
color = [155,155,155]

def closest(colors,color):
    colors = np.array(colors)
    color = np.array(color)
    distances = np.sqrt(np.sum((colors-color)**2,axis=1))
    index_of_smallest = np.where(distances==np.amin(distances))
    smallest_distance = colors[index_of_smallest]
    return (smallest_distance[0][0], smallest_distance[0][1], smallest_distance[0][2])

closest_color = closest(list_of_colors,color)
print(closest_color)
```

# Try the Color Finder on the 2-Cut Colors

```python
rgb_simple['cut_2_color_1'] = rgb_simple['cut_2'].str.extract(r'\((.*?)\)')
```

```python
rgb_simple['cut_2_color_1'] = rgb_simple['cut_2_color_1'].apply(lambda x: [int(i) for i in x.split(',')])
```

```python
import ast
# Convert string representation to list of tuples
rgb_simple['cut_2_list'] = rgb_simple['cut_2'].apply(ast.literal_eval)
# Extract the second tuple from each list
rgb_simple['cut_2_color_2'] = rgb_simple['cut_2_list'].apply(lambda x: list(map(int, x[1])) if x and len(x) > 1 else None)
```

```python
rgb_simple['closest_color_1'] = rgb_simple['cut_2_color_1'].apply(lambda x: closest(list_of_colors, x))
```

```python
rgb_simple['closest_color_2'] = rgb_simple['cut_2_color_2'].apply(lambda x: closest(list_of_colors, x))
```

```python
del rgb_simple['cut_2_list']
```

```python
rgb_simple[['Title', 'Artist', 'cut_2', 'closest_color_1', 'closest_color_2']].head(25)
```

# Cut 4

```python
cut4 = rgb_simple[['Title', 'Artist', 'Date', 'cut_4']]
```

```python
cut4.head()
```

```python
# Convert string representation to list of tuples
cut4['cut_4_list'] = cut4['cut_4'].apply(ast.literal_eval)
```

```python
for i in range(4):
    cut4_color = f'cut4_color_{i+1}'
    cut4[cut4_color] = cut4['cut_4_list'].apply(lambda x: list(map(int, x[i])) if x and len(x) > 1 else None)
```

```python
for i in range(4):
    cut4_color = f'cut4_color_{i+1}'
    closest_color = f'closest_color_{i+1}'
    cut4[closest_color] = cut4[cut4_color].apply(lambda x: closest(list_of_colors, x))
```

```python
cut4[['Date', 'closest_color_1', 'closest_color_2', 'closest_color_3', 'closest_color_4']].head(10)
```

```python
melted_df = pd.melt(cut4, id_vars=['Date'], value_vars=['closest_color_1', 'closest_color_2', 'closest_color_3', 'closest_color_4'], value_name='Color')

# Drop the 'variable' column
melted_df.drop(columns=['variable'], inplace=True)

# Group by 'Date' and 'Color' and count occurrences
grouped_df = melted_df.groupby(['Date', 'Color']).size().reset_index(name='Count')

grouped_df.sort_values(['Date', 'Count'], ascending=[True, False], inplace=True)
grouped_df.reset_index(inplace=True)
del grouped_df['index']
```
