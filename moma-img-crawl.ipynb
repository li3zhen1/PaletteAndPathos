```python
# read artworks_painting.csv

import pandas as pd

df = pd.read_csv('artworks_painting.csv')
df.head()
```

```python
import os
import time
from selenium import webdriver

# from "Thumbnail URL" column
image_urls = df['ThumbnailURL'].tolist()

# Set up the WebDriver (replace 'YOUR_PATH' with the path to your WebDriver)
driver = webdriver.Chrome()

# Folder to save images
save_folder = "painting"
if not os.path.exists(save_folder):
    os.makedirs(save_folder)
error_list = []
# Iterate over the df
for row in df.itertuples():
    url = row.ThumbnailURL
    objectID = row.ObjectID
    if os.path.exists(save_folder + '/' + str(objectID) + '.png'):
        continue
    try:
        driver.get(url)
        
        # Wait for the page to load (you can adjust the time as necessary)
        time.sleep(1)

        # save the image on the page
        image = driver.find_element("xpath", '(//img)[1]')
        # right click on the image and save as "<ObjectID>.png"
        image.screenshot(save_folder + '/' + str(objectID) + '.png')
        time.sleep(0.2)
    except:
        print(url)
        print("Error saving image for ObjectID: " + str(objectID))
        error_list.append(objectID)



    

driver.quit()

#save error list
error_df = pd.DataFrame(error_list, columns=['ObjectID'])
error_df.to_csv('error_list.csv', index=False)
```
