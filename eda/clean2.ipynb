```python
import pandas as pd


WIKI_CLASS_PATH = "/Volumes/T7/wikiart/wikiart/classes.csv"
```

```python
original_class_df = pd.read_csv(WIKI_CLASS_PATH)

original_class_df.head()
```

```python
original_class_df["Folder_Name"] = original_class_df["filename"].str.split("/", expand=True)[0]
original_class_df["File_Name"] = original_class_df["filename"].str.split("/", expand=True)[1]
```

```python
original_class_df.head()
```

```python
WIKI_I2E = '../results/image2emotion/wikiart'
import os
i2e_dict = {

}

for cate in os.listdir(WIKI_I2E):
    i2e_dict[cate.split('.')[0]] = pd.read_csv(os.path.join(WIKI_I2E, cate))

i2e_dict
```

```python
# for each row in original_class_df, find the corresponding emotion

def find_emotion(row):
    folder_name = row["Folder_Name"]
    file_name = row["File_Name"].split(".")[0]
    emotion_df = i2e_dict[folder_name]
    emotion_row = emotion_df[emotion_df["ObjectID"] == file_name]
    return emotion_row
```

```python
find_emotion(original_class_df.iloc[0])
```

```python
# add columns to original_class_df: amusement	awe	contentment	excitement	anger	disgust	fear	sadness	something else
import numpy as np

original_class_df["amusement"] = np.nan
original_class_df["awe"] = np.nan
original_class_df["contentment"] = np.nan
original_class_df["excitement"] = np.nan
original_class_df["anger"] = np.nan
original_class_df["disgust"] = np.nan
original_class_df["fear"] = np.nan
original_class_df["sadness"] = np.nan
original_class_df["something else"] = np.nan

# for each row in original_class_df, find the corresponding emotion, and add the emotion to the row
for index, row in original_class_df.iterrows():
    emotion_row = find_emotion(row)
    if emotion_row.empty:
        print(index, row)
    else:
        original_class_df.loc[index, "amusement"] = emotion_row["amusement"].values[0]
        original_class_df.loc[index, "awe"] = emotion_row["awe"].values[0]
        original_class_df.loc[index, "contentment"] = emotion_row["contentment"].values[0]
        original_class_df.loc[index, "excitement"] = emotion_row["excitement"].values[0]
        original_class_df.loc[index, "anger"] = emotion_row["anger"].values[0]
        original_class_df.loc[index, "disgust"] = emotion_row["disgust"].values[0]
        original_class_df.loc[index, "fear"] = emotion_row["fear"].values[0]
        original_class_df.loc[index, "sadness"] = emotion_row["sadness"].values[0]
        original_class_df.loc[index, "something else"] = emotion_row["something else"].values[0]
```

```python
# save to csv

original_class_df.to_csv("../cleaned_data/wikiart_with_emotion.csv", index=False)
```

```python

cutDict = {}
CUT_PATH = "../results/color_cut/wikiart"

for cate in os.listdir(CUT_PATH):
    if cate.endswith("_cut_hex.csv"):
        cutDict[cate.split('_cut_hex.')[0]] = pd.read_csv(os.path.join(CUT_PATH, cate))

cutDict
        
```

```python
col_names = []
for cut_count in [2, 4, 8]:
    for i in range(cut_count):
        col_name = f"cutBy{cut_count}_{i}"
        col_names.append(col_name)

col_names
```

```python
for col_n in col_names:
    original_class_df[col_n] = np.nan

def find_cut(row):
    folder_name = row["Folder_Name"]
    file_name = row["File_Name"]
    cut_df = cutDict[folder_name]
    cut_row = cut_df[cut_df["ObjectID"] == file_name]
    return cut_row

for index, row in original_class_df.iterrows():
    emotion_row = find_cut(row)
    if emotion_row.empty:
        print(index, row)
    else:
        for col_n in col_names:
            original_class_df.loc[index, col_n] = emotion_row[col_n].values[0]
```
